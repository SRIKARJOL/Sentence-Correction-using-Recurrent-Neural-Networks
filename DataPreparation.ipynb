Our dataset has around 2000 SMS messages represented in standard English,Chinese and SMS format. We will only be taking the 
English and SMS format from it. This notebook is used to prepare the data by storing the SMS texts in one column and standard 
English texts in another column.

Sample Data - 

image.png

```python
import pandas as pd
```

```python
# Opening a file and reading all the lines
file_name = "/content/data.txt"
file=open(file_name,'r')
data=file.readlines()
file.close()
```

```python
len(data)
```

```python
# To get only every first and second text in a group of 3 type of texts
sms_text = []
english_text = []
for index,each_line in enumerate(data):
  if index % 3 == 0:
    sms_text.append(each_line.strip())
  elif index % 3 == 1:
    english_text.append(each_line.strip())
  else:
    pass  
```

```python
print(len(sms_text))
print(len(english_text))
```

```python
print(sms_text[0])
print(english_text[0])
```

```python
# Creating dataframe
data_dict={'SMS_TEXT':sms_text,'ENGLISH_TEXT': english_text}
df = pd.DataFrame(data_dict)
df.head()
```

```python
df.to_csv("prepared_data.csv",index=False)
```
